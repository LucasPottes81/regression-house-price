```python
import pandas as pd
import numpy as np


df_csv1 = pd.read_csv('../data/raw/sample_submission.csv')
df_csv2 = pd.read_csv('../data/raw/train.csv')
df_csv3 = pd.read_csv('../data/raw/test.csv')

pd.set_option('display.max_columns', None)

display(df_csv1.head())
display(df_csv2.head())
display(df_csv3.head())

display(df_csv1.info())
display(df_csv2.info())
display(df_csv3.info())

display(df_csv1.describe())
display(df_csv2.describe())
display(df_csv3.describe())
```

```python
print(df_csv1.isnull().sum())
print(df_csv2.isnull().sum())
print(df_csv3.isnull().sum())
```

```python
df_csv1.fillna(df_csv1.mean(numeric_only=True), inplace=True)
df_csv2.fillna(df_csv2.mean(numeric_only=True), inplace=True)
df_csv3.fillna(df_csv3.mean(numeric_only=True), inplace=True)


df_csv1.drop_duplicates(inplace=True)
df_csv2.drop_duplicates(inplace=True)
df_csv3.drop_duplicates(inplace=True)
```

```python
df_csv1.collumns = df_csv1.columns.str.lower().str.replace(' ', '_')
df_csv2.collumns = df_csv2.columns.str.lower().str.replace(' ', '_')
df_csv3.collumns = df_csv3.columns.str.lower().str.replace(' ', '_')
```

```python
display(df_csv1.describe())
display(df_csv2.describe())
display(df_csv3.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Quais variáveis mais impactam o preço?

```python
colunas_numericas = df_csv2.select_dtypes(include=[np.number]).columns.tolist()

correlacao = df_csv2[colunas_numericas].corr()

# Pegamos as 11 primeiras porque a 1ª da lista será o próprio SalePrice 
top_10_vars = correlacao['SalePrice'].abs().sort_values(ascending=False).head(10).index

matriz_top10 = df_csv2[top_10_vars].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_top10, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Matriz de Correlação - Top 10 Variáveis')

print("Ranking das 10 variáveis mais correlacionadas com SalePrice:")
print(correlacao['SalePrice'].abs().sort_values(ascending=False).head(10))

```

## Área construída tem relação linear com preço?


```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df_csv2, x='GrLivArea', y='SalePrice', color='blue', alpha=0.6)

sns.regplot(data=df_csv2, x='GrLivArea', y='SalePrice', scatter=False, color='red')

plt.title('Relação entre GrLivArea e SalePrice', fontsize=15)
plt.xlabel('Área do Living Room (sq ft)', fontsize=12)
plt.ylabel('Preço de Venda (USD)', fontsize=12)

sns.despine()
plt.show()
```

# Ano de construção influencia valorização?

```python
preco_por_ano = df_csv2.groupby('YearBuilt')['SalePrice'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=preco_por_ano, x='YearBuilt', y='SalePrice', color = 'darkblue', linewidth=2.5)

plt.title('Evolução do Preço Mediano pelo Ano de Construção', fontsize=16)
plt.xlabel('Ano de Construção', fontsize=12)
plt.ylabel('Preço Mediano de Venda ($)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)
sns.despine()

plt.show()
```

#  Casas reformadas valem mais?

```python
df_csv2['Foi_Reformada'] = np.where(df_csv2['YearRemodAdd'] > df_csv2['YearBuilt'], 'Sim', 'Não')

plt.figure(figsize=(8, 5))

sns.boxplot(data=df_csv2, x='Foi_Reformada', y='SalePrice', palette=['lightcoral', 'lightblue'])

plt.title('Impacto da Reforma no Preço de Venda', fontsize=15)
plt.xlabel('A casa foi reformada?', fontsize=12)
plt.ylabel('Preço de Venda ($)', fontsize=12)
sns.despine()

plt.show()

medianas = df_csv2.groupby('Foi_Reformada')['SalePrice'].median()

print(f"Casas ORIGINAIS (Não reformadas): ${medianas['Não']:,.2f}")
print(f"Casas REFORMADAS: ${medianas['Sim']:,.2f}")
```

# Bairros apresentam diferenças estatísticas relevantes?

```python
import scipy.stats as stats

ordem_bairros = df_csv2.groupby('Neighborhood')['SalePrice'].median().sort_values(ascending=False).index

plt.figure(figsize=(12, 6))

sns.boxplot(data=df_csv2, x='Neighborhood', y='SalePrice', order=ordem_bairros, palette='coolwarm')

plt.title('Preço de Venda por Bairro (Do mais caro ao mais barato)', fontsize=16)
plt.xlabel('Bairro (Neighborhood)', fontsize=12)
plt.ylabel('Preço de Venda ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
sns.despine()


plt.show()

# Teste anova (realizar a média de 3 ou mais grupos) para verificar se há diferença significativa entre os bairros em relação ao preço de venda dos imóveis.
grupos_bairros = [df_csv2[df_csv2['Neighborhood'] == bairro]['SalePrice'] for bairro in df_csv2['Neighborhood'].unique()]

f_stat, p_valor = stats.f_oneway(*grupos_bairros)

print("\n--- Resultado do Teste Estatístico (ANOVA) ---")
print(f"Valor-P (p-value): {p_valor}")

# Interpretando o resultado para o negócio
if p_valor < 0.05:
    print("Veredito: SIM! O bairro apresenta uma diferença ESTATISTICAMENTE RELEVANTE no preço do imóvel.")
else:
    print("Veredito: NÃO. O bairro não influencia estatisticamente no preço.")
```
